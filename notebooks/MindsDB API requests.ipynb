```python
from env import config
import requests
```

```python
# connect
url = 'http://127.0.0.1:47334/api/sql/query'

# query
resp = requests.post(url, json={'query':
                    'SELECT m.Datetime as Datetime, m.Open as forecasted_open FROM mindsdb.btc_price_predictor as m JOIN btc_datasource.btc_price_data as t;'})

# response
print(resp.json())
```

```python
import json

response_data = resp.json()

column_names = response_data['column_names']
data = response_data['data']
datetime_index = column_names.index('Datetime')
price_index = column_names.index('forecasted_open')

result_dict = {entry[datetime_index]: entry[price_index] for entry in data}
result_json = json.dumps(result_dict, indent=2)
print(result_json)
```
